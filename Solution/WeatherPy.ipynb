```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy


# Output File (CSV)
output_data_file = "Output/weather_df.csv"


# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)


#print api_key
#api_key

```

## Generate Cities List

```python

# List for holding lat_lngs and cities
lat_lngs = []
cities = []


# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)


# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

        
# Print the city count to confirm sufficient count
len(cities)
#cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key

```

```python

# Creating empty lists to append the API data into lists 

city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []


# Log file print statement
print(f"Beginning Data Retrieval")
print(f"-------------------------------")

# set initial call counter
record = 0
set = 1

# Loop through the cities in the city list 
for city in cities:  
    
    # Try statement is used to append calls where value is found 
    # Not all calls return data as OpenWeatherMap will not have records in all the cities generated by CityPy module
    
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        #print(response)
        city_name.append(response["name"])
        #print(city_name)
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        #print(f"Response Record respone ")
        #print(f"{url}&q={city}")
        
            
        if record > 49:
                record = 1
                set = set + 1

        else:
            record = record + 1
        
        print(f"Processing Record {record} of Set {set} | {city_record}")
        

         
    # If no record found "skip" to next call
    except:
        print("City not found. Skipping...")
    continue

print("----------------------------------")
print("Data Retrieval Complete")
print("----------------------------------")



    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python

# create a dictionary for establishing dataframe

weather_dict = {
    "City": city_name,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Lat": lat,
    "Lng": lng,
    "Max Temp": max_temp,
    "Wind Speed": wind_speed
}
```

```python
# establish dataframe
#weather_df = pd.DataFrame.from_dict(weather_dict)

weather_df = pd.DataFrame(weather_dict)
weather_df.head()
weather_df
```

```python
# Output data to csv file "weather_df.csv"

weather_df.to_csv(output_data_file)
```

## Data Analysis based on WeatherPy


By using WeatherPy we are considering the weather of 500+ cities across the world of varying distance from the equator.The WeatherPy shows relationship between the "temperature and city latitude", "humidity and city latitude", "cloudiness and city latitude", "wind speed and city latitude".The observable trends based on the data are:-


From the analysis we can conclude that the temperature have a significant correlation with the latitude of the cities.The temperatures are higher, the closer they are to the equator (that is zero degree latitude).The weather becomes significantly warmer as one approaches the equator.Majority of cities are located at latitude greater than -40 degree and those cities have temperatue below 90 degree F.

There is no specific relation between latitude of cities and humidity. However there is a slightly larger cluster of cities are present in the high humidity region(above 40% humidity).


There is no specific relation between latitude of cities and cloudiness.Cloudiness is consistent across all latitueds. However, it is interesting to see that a strong band of cities near 0, 75, and 100 % cloudiness.Based on the data, it seems as though cloudiness and humidity are relatively consistent across all latitudes. 


There is no specific relation between latitude of cities and wind speed.But cities are mostly located in the region with low wind speed (below 15mph).Wind speed tends to generally be betweeen 0 and 15 mph regardless of latitude.
